```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import os
from os import listdir
from os.path import isfile, join
```

```python
DATA_PATH = 'img'
LISTED_DIR = list(set(os.listdir(DATA_PATH))-{'desktop.ini'})
```

```python
images = [f for f in LISTED_DIR if isfile(join(DATA_PATH, f))]
```

```python
STARTING_HEIGHT = 960
STARTING_WIDTH = 585

PRODUCTION_HEIGHT = 43
PRODUCTION_WIDTH = 65

count=0
for i in images:
    print(i)
    img = mpimg.imread(join(DATA_PATH, i))
    #img.show()
    imgplot = plt.imshow(img[STARTING_HEIGHT:STARTING_HEIGHT+PRODUCTION_HEIGHT, STARTING_WIDTH:STARTING_WIDTH+PRODUCTION_WIDTH])
    plt.show()
    #plt.imsave('d'+str(count)+'.png', img[STARTING_HEIGHT:STARTING_HEIGHT+PRODUCTION_HEIGHT, STARTING_WIDTH:STARTING_WIDTH+PRODUCTION_WIDTH])
    count+=1
```
