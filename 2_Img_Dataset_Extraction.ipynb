```python
import os
from PIL import Image

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from os import listdir
from os.path import isfile, join
import pathlib
```

```python
DATA_PATH = 'Img_to_Slice'
LISTED_DIR = list(set(os.listdir(DATA_PATH))-{'desktop.ini'})
```

```python
images = [f for f in LISTED_DIR if isfile(join(DATA_PATH, f))]
```

```python
#You might want to adjust the positon, depending on what source you're getting the images from.

#Setting the production portrait slices:
STARTING_HEIGHT = 60
STARTING_WIDTH = 55

PRODUCTION_HEIGHT = 50
PRODUCTION_WIDTH = 47



count=0
NEXT_PRODUCTION= 16 + PRODUCTION_WIDTH
for i in images:
    print(i)
    img = mpimg.imread(join(DATA_PATH, i))
    #img = Image.open(join(DATA_PATH, i))
    for j in range(9):
        #This moves 9 squared portraits horizontally, it normally gets noisy data, but it'll hardly let late game's production tabs pass
        starting_width = STARTING_WIDTH + NEXT_PRODUCTION * j
        
        imgplot = plt.imshow(img[STARTING_HEIGHT:STARTING_HEIGHT+PRODUCTION_HEIGHT, starting_width:starting_width+PRODUCTION_WIDTH])
        plt.show()
        
        plt.imsave('Sliced_ProductionBar_Imgs/d'+str(count)+'.png', img[STARTING_HEIGHT:STARTING_HEIGHT+PRODUCTION_HEIGHT,
                                                                        starting_width:starting_width+PRODUCTION_WIDTH])
        #The d at the beginning is to avoid name repetiton of the images to label.
        count+=1
```
