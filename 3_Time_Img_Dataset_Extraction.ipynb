```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import os
from os import listdir
from os.path import isfile, join
```

```python
DATA_PATH = 'Img_to_Slice'
LISTED_DIR = list(set(os.listdir(DATA_PATH))-{'desktop.ini'})
```

```python
images = [f for f in LISTED_DIR if isfile(join(DATA_PATH, f))]
```

```python
STARTING_HEIGHT = 770
STARTING_WIDTH = 0

PRODUCTION_HEIGHT = 28
PRODUCTION_WIDTH = 35

count=0
for i in images:
    print(i)
    img = mpimg.imread(join(DATA_PATH, i))
    starting_width = STARTING_WIDTH
    #img.show()
    imgplot = plt.imshow(img[STARTING_HEIGHT:STARTING_HEIGHT+PRODUCTION_HEIGHT, STARTING_WIDTH:STARTING_WIDTH+PRODUCTION_WIDTH])
    plt.show()
    plt.imsave('Sliced_Time_Img/b'+str(count)+'.png', img[STARTING_HEIGHT:STARTING_HEIGHT+PRODUCTION_HEIGHT, STARTING_WIDTH:STARTING_WIDTH+PRODUCTION_WIDTH])
    #Again, b on the beginning of the file's name is due to avoid name repetition.
    count+=1
```
